```python
# Carregar os dados das ações 
import yfinance as yf 
```

```python
ticker = input("digite o código da ação: ")
dados = yf.Ticker(ticker).history("1y")
dados.head()
```

```python
treinamento = dados.reset_index()
treinamento 
```

```python
# Selecionando duas ou mais colunas 
treinamento = treinamento[['Date', 'Close']]
treinamento
```

```python
treinamento['Date'].dt.tz_localize(None)
```

```python
treinamento['Date'] = treinamento ['Date'].dt.tz_localize(None)
treinamento
```

```python
# Renomeando o nome das colunas 
treinamento.columns = ['ds', 'y']
treinamento
```

```python
# Treinando o modelo de machine learning 
# Instalando a Ia do facebook 
!pip install prophet
```

```python
from prophet import Prophet 
from prophet.plot import plot_plotly
```

```python
# Criar modelo IA
modelo = Prophet()
```

```python
# Treinar modelo 
modelo.fit(treinamento)
```

```python
# Realizar previsões
periodo = modelo.make_future_dataframe(90)
```

```python
periodo.tail(10)
```

```python
previsoes = modelo.predict(periodo)
previsoes
```

```python
# Gerar vizualição Gráfica 
plot_plotly(modelo, previsoes)
```
